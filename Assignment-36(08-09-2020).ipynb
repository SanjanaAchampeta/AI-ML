```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ig=plt.imread("tom.jpg")
print(ig.shape)
```

```python
plt.imshow(ig)
```

```python
plt.imshow(ig[:,:,0])
```

```python
plt.imshow(ig[:,:,0],cmap='gray')
```

```python
plt.imshow(ig[:,0:200])
```
